    - - ### Q2
      - MFCC: MFCC is one of the most commonly used features in speech processing. It is a set of cepstral coefficients extracted based on Mel-frequency bands, which are logarithmically spaced to match the human auditory system.
      - ZSC: This feature calculates the number of times a signal crosses zero in a specified time interval.
      - SM: Mel spectrogram is a two-dimensional representation of the frequency spectrum of speech, generated based on Mel-frequency bands.
      - CM: Chroma feature is related to the distribution of notes (or chords) in the audio spectrum.
        - - In many cases, combining these features can improve the accuracy and efficiency of speech recognition models
    - - ### Q3
      - Sensitivity to Frequency Changes: MFCCs are sensitive to changes in the frequencies present in the audio signal, especially those frequencies that are crucial for speech perception.
      - Robustness to Noise: MFCCs are generally robust to low-level noise and disturbances due to the use of logarithmic power spectrum, which reduces noise effects
      - Sensitivity to Temporal Changes :MFCCs are computed over short-time frames. Rapid changes in time, such as changes in speech rate, can significantly impact MFCC values.

- ## HHM
    - - ### Q1
      - States refer to concealed variables within the system, while observations pertain to visible variables or evidence accessible to us.
      - In this CA, states represent numerical values ranging from 0 to 9, while observations are derived from audio signals. The process of obtaining observations entails extracting pertinent features from the audio signal, and in this case, we utilize MFCC for feature extraction.
      - ### Q2
      - It is known as the first-order because the current state depends only on the prior state.
      - The distinction among different categories lies in the number of previous states upon which it depends. For instance, in a second-order model, the probability is contingent upon the two preceding states, and in a third-order model, it relies on the three preceding states.
    - - ### Q3
      - HMMs, are adept at modeling systems where the current state is solely influenced by the preceding state (following the Markov property). This characteristic makes them well-suited for representing processes characterized by inherent temporal structures. HMMs are especially valuable in scenarios where the underlying states are not directly observable but still impact the observed outcomes.
      - ### Q4
      - Advantages:
      - HMMs offer a probabilistic framework for modeling sequential data, allowing them to capture uncertainties and enhancing model robustness
      - They excel at capturing sequential dependencies in data, making them valuable for time-series data analysis
      - Disadvantages:
      - HMMs are sensitive to the choice of initial parameters, and incorrect initialization may lead to convergence to suboptimal solutions.
      - As the sequence length increases, HMMs tend to forget past states, which can limit their effectiveness in tasks requiring long-term memory
      - ### Q5
      - First-Order HMM: Transitions occur linearly from one state to the next, following the Markov property where the current state depends only on the previous state.
      - Hidden Semi-Markov Model (HSMM): HSMMs allow states to emit observations for variable durations, unlike traditional HMMs where each state emits exactly one observation
      - Left-to-right HMM transitions from one state to the next in a fixed order, useful for processes with sequential evolution.
