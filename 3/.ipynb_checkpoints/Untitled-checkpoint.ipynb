### In this section, we extract the features of each photo according to the vgg16 model (the pictures above).

```python
def VGGupdated(input_tensor=None,classes=10):    
   
    img_rows, img_cols = 224, 224 
    img_channels = 3

    img_dim = (img_rows, img_cols, img_channels)
   
    img_input = Input(shape=img_dim)
    
    # Block 1
    x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv1')(img_input)
    x = Conv2D(64, (3, 3), activation='relu', padding='same', name='block1_conv2')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block1_pool')(x)

    # Block 2
    x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv1')(x)
    x = Conv2D(128, (3, 3), activation='relu', padding='same', name='block2_conv2')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block2_pool')(x)

    # Block 3
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv1')(x)
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv2')(x)
    x = Conv2D(256, (3, 3), activation='relu', padding='same', name='block3_conv3')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block3_pool')(x)

    # Block 4
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv1')(x)
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv2')(x)
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block4_conv3')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block4_pool')(x)

    # Block 5
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv1')(x)
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv2')(x)
    x = Conv2D(512, (3, 3), activation='relu', padding='same', name='block5_conv3')(x)
    x = MaxPooling2D((2, 2), strides=(2, 2), name='block5_pool')(x)

  
    model = Model(inputs = img_input, outputs = x, name='VGGdemo')


    return model
```

## Reading and storing images 

in this section im going to read the images, store them in an array, and then pass them to the VGG16 model to extract features. Below is the code to achieve this <br>
Here, features is a data matrix where each row corresponds to one image. First, the data is reshaped into a matrix with dimensions (n_samples, height * width * channels) using the reshape function. This transformation converts each image into a vector of length height * width * channels.

```python
import os
def read_and_preprocess_images(directory_path):
    images = []
    for i in range(1, 211):
        filename = '{:04d}.png'.format(i)
        image_path = os.path.join(directory_path, filename)
        if os.path.exists(image_path):
            img = image.load_img(image_path, target_size=(224, 224))
            img_array = image.img_to_array(img)
            img_array = np.expand_dims(img_array, axis=0)
            img_array = preprocess_input(img_array)
            images.append(img_array)
        else:
            print(f"Image file {filename} not found.")
    
    return np.vstack(images)

# Read and preprocess images
directory_path = "flower_images"
images_array = read_and_preprocess_images(directory_path)
len(images_array)

model = VGGupdated(classes = 10)
features = model.predict(images_array)
features
```

## Running k-means
in the following function im going to run the k-means algorithm  for analyzing image data. <br>
as i mentioned in previous questions i used 10 cluster for the algorithm (n_clusters = 10)

```python

n_samples, height, width, channels = features.shape
features_reshaped = features.reshape(n_samples, height * width * channels)

n_clusters = 10

kmeans = KMeans(n_clusters=n_clusters,n_init=10)
kmeans.fit(features_reshaped)
cluster_labels = kmeans.labels_
```

Finally, the clustering labels for each image are returned as cluster_labels, where each label corresponds to one of the n_clusters clusters that each image belongs to.

```python
cluster_labels
```

## Dimension reduction (pca)
### Q7
- pca is a dimensionality reduction method commonly used for data analysis, particularly for reducing the dimensions of feature space. It helps us effectively decompose dense and redundant information present in high-dimensional data into a more manageable and meaningful form.


in the code below we are going to create a scatter plot to visualize the clusters in the reduced dimensional space. It loops through each cluster and plots the points corresponding to that cluster label. It uses boolean indexing to select points belonging to each cluster label. Finally, it adds labels, title, and legend to the plot, then displays it.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
features_pca = pca.fit_transform(features_reshaped)

# Plot clusters in the reduced space
plt.figure(figsize=(10, 6))
for cluster_label in range(n_clusters):
    plt.scatter(features_pca[cluster_labels == cluster_label, 0], 
                features_pca[cluster_labels == cluster_label, 1], 
                label=f'Cluster {cluster_label}')
plt.title('Clusters in Reduced Dimensional Space')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Apply PCA for dimensionality reduction
pca = PCA(n_components=2)  # Reduce to 2 dimensions for visualization
features_pca = pca.fit_transform(features_reshaped)

# Apply DBSCAN clustering on reduced features
dbscan = DBSCAN(eps=0.3, min_samples=10)  # You may need to adjust the parameters based on your data
cluster_labels = dbscan.fit_predict(features_pca)

# Plot clusters in the reduced space
plt.figure(figsize=(10, 6))
unique_labels = set(cluster_labels)
for label in unique_labels:
    if label == -1:
        # Noise points are plotted as black crosses
        plt.scatter(features_pca[cluster_labels == label, 0], 
                    features_pca[cluster_labels == label, 1], 
                    c='k', marker='x', label='Noise')
    else:
        plt.scatter(features_pca[cluster_labels == label, 0], 
                    features_pca[cluster_labels == label, 1], 
                    label=f'Cluster {label}')
plt.title('Clusters in Reduced Dimensional Space (DBSCAN)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

### Q8
- Homogeneity:
- - To calculate this metric, we first compare the actual data labels and the clusters that the clustering algorithm created.
If a cluster includes only members of one class, the homogeneity criterion of that cluster is 1. But if the data of one class are located in several clusters, the homogeneity criterion will be less than 1.
This measure is useful for cases where real labels are available, as we can compare the performance of the clustering algorithm with real labels. are available.
- Silhouette Score:
- - This score indicates how similar an object is to its own cluster (integration) compared to other clusters (separation). The silhouette score is in the range of -1 to 1, where a higher value means a good match of the object with its own cluster and also a lack of match with the neighboring clusters.

```python
from sklearn.metrics import homogeneity_score, silhouette_score

kmeans = KMeans(n_clusters=n_clusters)
kmeans_labels = kmeans.fit_predict(features_reshaped)

dbscan = DBSCAN(eps=0.3, min_samples=10)
dbscan_labels = dbscan.fit_predict(features_reshaped)
kmeans_homogeneity = homogeneity_score(df, kmeans_labels)

kmeans_silhouette = silhouette_score(features_reshaped, kmeans_labels)

dbscan_silhouette = silhouette_score(features_reshaped, dbscan_labels)

print("K-means Homogeneity Score:", kmeans_homogeneity)
print("K-means Silhouette Score:", kmeans_silhouette)
print("DBSCAN Silhouette Score:", dbscan_silhouette)
```
